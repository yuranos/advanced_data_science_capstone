### Broad look at the data
Let's just look at the data overall. We have 3000 rows and 18 columns(features).

```python
import pandas as pd

users = pd.read_csv('fitness_wearables.csv')

users.head(5)
```

### Compute labels
First thing I'd do is create a binary label as we don't it right now. To label every wearable user, I'd like to get an average of the first 3 weeks and the last three weeks(last 6 columns).
Let's call them 1m_average, 1y_average:

```python
def calculate_average(row):
    _1_m_average = (row['N sessions/w 1m1w'] + row['N sessions/w 1m2w'] + row['N sessions/w 1m3w'])/3
    _1_y_average = (row['N sessions/w 1y1w'] + row['N sessions/w 1y2w'] + row['N sessions/w 1y3w'])/3
    return (_1_m_average, _1_y_average)

users['_1_m_average'] = users.apply(lambda row: calculate_average(row)[0], axis=1)
users['_1_y_average'] = users.apply(lambda row: calculate_average(row)[1], axis=1)
users.head(10)
```

### Sticklers
Let's now think of a threshold. My assumption is that if the ratio between the first and the second average(1 month vs 1 year) is >=75%, it's a good retention(**stickler**), and we can say that a person is training at the same rate as when they bought a wearable.

On the other hand, anything that's <75% - is a bad retention meaning that the person is not training as hard as they used to in the beginning.

```python
def is_stickler(row):
    return True if (row['_1_y_average'] / row['_1_m_average']) >= 0.75 else False 
users['stickler'] = users.apply(lambda row: is_stickler(row), axis=1)
users.head(10)
```
